```python
df = pd.read_csv('FuelConsumption.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#Piechart
fueltype = df['FUELTYPE'].value_counts()
fueltype.plot(kind='pie',autopct='%1.1f%%', wedgeprops={'edgecolor': 'black'})
plt.title("Fueltype Distribution")
plt.legend()
plt.show()
```

```python
Vehicleclass = df['VEHICLECLASS'].value_counts()
Vehicleclass.plot(kind='pie',autopct='%1.1f%%', wedgeprops={'edgecolor': 'turquoise'})
plt.title("Vehicleclass Distribution")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=df.index, y='CO2EMISSIONS', data=df)
plt.title('CO2 Emissions Readings Across All Entries')
plt.xlabel('Entry Index')
plt.ylabel('CO2 Emissions')
plt.grid(True)
plt.show()
```
